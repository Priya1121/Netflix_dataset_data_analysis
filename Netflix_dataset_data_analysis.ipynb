### 6. In which year the highest no. of TV shows and movies were relaesed ? Also plot the bar graph.

```python
y.index   # the index of the y series is the year 
```

```python
plt.bar(y.index,y,edgecolor='black')   # here we are ploting the graph between the year and no. of shows released in it.
plt.xlabel('Year')
plt.ylabel('No. of show released')
plt.title('Graph between "Year" and "No. of show released"')
```

Result : In year 2019 , 2154 shows had released which is the highest number.
